```python
import math
import numpy as np
import pandas as pd
import random as rd
```

```python
class carta:
    def __init__(self, numero=10, palo="Oro"):
        self.numero = numero
        self.palo = palo
    
    def __str__(self):
        if self.numero==10: numstr="As"
        if self.numero==9: numstr="3"
        if self.numero==8: numstr="Rey"
        if self.numero==7: numstr="Caballo"
        if self.numero==6: numstr="Sota"
        if self.numero==5: numstr="7"
        if self.numero==4: numstr="6"
        if self.numero==3: numstr="5"
        if self.numero==2: numstr="4"
        if self.numero==1: numstr="2"
        if self.numero==0: numstr="0"
        return f"{numstr} de {self.palo}s"
    
    def __eq__(self, other):
        return self.numero == other.numero and self.palo == other.palo
    
    def evalua(cartemp, palpinta, palprimero):
        valor = cartemp.numero
        if cartemp.palo == palprimero and palprimero != palpinta:
            valor += 10
        if cartemp.palo == palpinta:
            valor += 20
        return valor
```

```python
def decide(palpinta, diferenciac, primerac, ganandoc, mano, manualdecide):
    icd = list()
    icdmax = list()
    icdmin = list()
    cdmin = 100
    cdmax = 0
    
    if manualdecide > -1:
        return manualdecide
    
    if primerac is None:
        for nm in range(len(mano)):
            if mano[nm].evalua(palpinta, None) >= cdmax:
                cdmax = mano[nm].evalua(palpinta, None)
                icdmax.append(nm)
            if mano[nm].evalua(palpinta, None) <= cdmin:
                cdmin = mano[nm].evalua(palpinta, None)
                icdmin.append(nm)
        if diferenciac > 0:
            return rd.choice(icdmax)
        elif diferenciac < 0:
            return rd.choice(icdmin)
        return rd.randint(0, len(mano)-1)
    
    if ganandoc.palo == primerac.palo:
        
        for nm in range(len(mano)):
            if mano[nm].palo == ganandoc.palo and mano[nm].numero > ganandoc.numero:
                if mano[nm].numero >= cdmax:
                    cdmax = mano[nm].numero
                    icdmax.append(nm)
                if mano[nm].numero <= cdmin:
                    cdmin = mano[nm].numero
                    icdmin.append(nm)
                icd.append(nm)
        if len(icd) > 0:
            if diferenciac > 0:
                return rd.choice(icdmax)
            elif diferenciac < 0:
                return rd.choice(icdmin)
            return rd.choice(icd)
        
        for nm in range(len(mano)):
            if mano[nm].palo == ganandoc.palo:
                if mano[nm].numero >= cdmax:
                    cdmax = mano[nm].numero
                    icdmax.append(nm)
                if mano[nm].numero <= cdmin:
                    cdmin = mano[nm].numero
                    icdmin.append(nm)
                icd.append(nm)
        if len(icd) > 0:
            if diferenciac > 0:
                return rd.choice(icdmax)
            elif diferenciac < 0:
                return rd.choice(icdmin)
            return rd.choice(icd)
        
        for nm in range(len(mano)):
            if mano[nm].palo == palpinta:
                if mano[nm].numero >= cdmax:
                    cdmax = mano[nm].numero
                    icdmax.append(nm)
                if mano[nm].numero <= cdmin:
                    cdmin = mano[nm].numero
                    icdmin.append(nm)
                icd.append(nm)
        if len(icd) > 0:
            if diferenciac > 0:
                return rd.choice(icdmax)
            elif diferenciac < 0:
                return rd.choice(icdmin)
            return rd.choice(icd)
        
        for nm in range(len(mano)):
            if mano[nm].evalua(palpinta, None) >= cdmax:
                cdmax = mano[nm].evalua(palpinta, None)
                icdmax.append(nm)
            if mano[nm].evalua(palpinta, None) <= cdmin:
                cdmin = mano[nm].evalua(palpinta, None)
                icdmin.append(nm)
        if diferenciac > 0:
            return rd.choice(icdmax)
        elif diferenciac < 0:
            return rd.choice(icdmin)
        return rd.randint(0, len(mano)-1)
    
    if ganandoc.palo == palpinta:
        
        for nm in range(len(mano)):
            if mano[nm].palo == primerac.palo:
                if mano[nm].numero >= cdmax:
                    cdmax = mano[nm].numero
                    icdmax.append(nm)
                if mano[nm].numero <= cdmin:
                    cdmin = mano[nm].numero
                    icdmin.append(nm)
                icd.append(nm)
        if len(icd) > 0:
            if diferenciac > 0:
                return rd.choice(icdmax)
            elif diferenciac < 0:
                return rd.choice(icdmin)
            return rd.choice(icd)
        
        for nm in range(len(mano)):
            if mano[nm].palo == palpinta and mano[nm].numero > ganandoc.numero:
                if mano[nm].numero >= cdmax:
                    cdmax = mano[nm].numero
                    icdmax.append(nm)
                if mano[nm].numero <= cdmin:
                    cdmin = mano[nm].numero
                    icdmin.append(nm)
                icd.append(nm)
        if len(icd) > 0:
            if diferenciac > 0:
                return rd.choice(icdmax)
            elif diferenciac < 0:
                return rd.choice(icdmin)
            return rd.choice(icd)
        
        for nm in range(len(mano)):
            if mano[nm].evalua(palpinta, None) >= cdmax:
                cdmax = mano[nm].evalua(palpinta, None)
                icdmax.append(nm)
            if mano[nm].evalua(palpinta, None) <= cdmin:
                cdmin = mano[nm].evalua(palpinta, None)
                icdmin.append(nm)
        if diferenciac > 0:
            return rd.choice(icdmax)
        elif diferenciac < 0:
            return rd.choice(icdmin)
        return rd.randint(0, len(mano)-1)
```

```python
palos = ["Oro", "Copa", "Espada", "Basto"]
```

```python
lbaraja = 40

njugadores = 5
nmano = 8

pinta = carta(0, palos[0])
manoini = [carta(8, palos[0]), carta(3, palos[0]), carta(1, palos[0]), carta(10, palos[1]), carta(5, palos[1]), carta(6, palos[2]), carta(5, palos[2]), carta(1, palos[3])]
empieza = 3
difc = -1

NMC = 100000

print("# de jugadores: "+str(njugadores))
print("Mano #: "+str(nmano))
print("Sale el jugador: "+str(empieza))
print("# de cartas - # de apuestas: "+str(difc))
print("# de MC: "+str(NMC))
print()

print("Usuario (Jugador 0):")
for nm in range(nmano):
    print(manoini[nm])
print()

print("Pinta:")
print(pinta)
print()
print()

barajan = list(range(lbaraja))

for md in range(nmano):
    histg = [[0 for j in range(njugadores)] for i in range(nmano+1)]
    for nMC in range(NMC):
        
        baraja = list()
        #print(barajan)
        rd.shuffle(barajan)
        #print(barajan)

        #print()
        for ic in range(njugadores*nmano+1):
                if ic > lbaraja-1: break
                #print(baraja[ic])
                baraja.append(carta(barajan[ic]%10+1, palos[math.floor(barajan[ic]/10)]))

        #print()
        #for c in baraja:
        #    print(c)

        # Crea las manos desconocidas a partir de las conocidas
        manos = [[carta() for j in range(nmano)] for i in range(njugadores)]
        manos[0][:] = manoini

        #print()
        ic=0
        while ic < len(baraja):
            if any(baraja[ic] == x for x in manos[0][:]) or baraja[ic] == pinta:
                #print(baraja[ic])
                del baraja[ic]
                ic-=1
            ic+=1

        #print()
        #for c in baraja:
        #    print(c)

        for nm in range(nmano):
            for nj in range(1,njugadores):
                manos[nj][nm] = baraja[(nj-1)+nm*(njugadores-1)]

#        print("Reparto:")
#        print()
#        for nj in range(njugadores):
#            print("Jugador "+str(nj+1)+":")
#            for nm in range(nmano):
#                print(manos[nj][nm])
#            print()
#    
#        print("Pinta:")
#        print(pinta)

        ig = [0 for j in range(nmano)]
        ict = [0 for i in range(njugadores)]
        ct = [[None for j in range(nmano)] for i in range(njugadores)]
        etemp = empieza
        for nm in range(nmano):
#            print()
#            print()
#            print("Mano "+str(nm+1)+":")
#            print()
            vg = 0
            njg = 0
            for nj in range(njugadores):
                if empieza == 0 and nm+nj == 0:
                    ict[nj] = decide(pinta.palo, difc, ct[0][nm], ct[njg][nm], manos[(etemp+nj)%njugadores][:], md)
                else:
                    ict[nj] = decide(pinta.palo, difc, ct[0][nm], ct[njg][nm], manos[(etemp+nj)%njugadores][:], -1)
                ct[nj][nm] = manos[(etemp+nj)%njugadores][ict[nj]]
                del manos[(etemp+nj)%njugadores][ict[nj]]
#                print("Jugador "+str((etemp+nj)%njugadores+1)+":")
#                print(ct[nj][nm])
                if ct[nj][nm].evalua(pinta.palo, ct[0][nm].palo) > vg:
                    vg = ct[nj][nm].evalua(pinta.palo, ct[0][nm].palo)
                    njg = nj
                    ig[nm] = (etemp+nj)%njugadores
            etemp = ig[nm]
#            print("Gana el jugador "+str(ig[nm]+1))
#        print()
#        print()
#        print("Resultado:")
#        print()
        for nj in range(njugadores):
            histg[ig.count(nj)][nj]+=1
#            print("Jugador "+str(nj+1)+": "+str(ig.count(nj)))
#        print()
#        print()
    for i in range(nmano+1):
        for j in range(njugadores):
            histg[i][j]/=NMC
    
    pmed = [[0 for j in range(njugadores)]]
    for nj in range(njugadores):
        for nm in range(nmano+1):
            pmed[0][nj] += nm*histg[nm][nj]
    
    histg = pd.DataFrame(histg, columns = range(njugadores), index=range(nmano+1))
    pmed = pd.DataFrame(pmed, columns = range(njugadores), index = ['pmed'])
    if empieza==0: print("Starting with card "+str(md+1))
    print(pd.concat([histg,pmed]))
    histg.plot(kind='bar', figsize=(10, 5), fontsize=18, ylim=[0,1], title="Starting with card "+str(md+1));
    print()
    print()
    if empieza!=0: break
```
